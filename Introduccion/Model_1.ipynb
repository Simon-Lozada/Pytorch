## Tareas

**1) Diseñar nuestro modelo:** (input, output, size, forward pass)

**2) Construir "loss"(perdida) y "optimizer"(optimizador)**

**3) Bucle de entrenamiento:**
 - Forward pass: computar predicción
 - backward pass: gradientes
 - Actulizar los pesos 

```python
import torch
import torch.nn as nn
```

```python

# f = w * x

# f = w * x
X = torch.tensor([[1], [2], [3], [4]], dtype = torch.float32)
Y = torch.tensor([[2], [4], [6], [8]], dtype = torch.float32)

X_test = torch.tensor([5], dtype = torch.float32)

n_samples, n_features = X.shape

print(n_samples, n_features)

input_size = n_features
output_size = n_features

# Podemos usar una capa simple del modulo "nn": model = nn.Linear(input_size, output_size)
# O Podemos hacer una modelo personalizado usando las capas del modulo "nn"
class LinealRegression(nn.Module):
    
    def __init__(self, input_dim, output_dim):
        #Defnimos a "LinealRegression" como una super clase
        super(LinealRegression, self).__init__()
        
        #Definir nuestras capas
        self.lin = nn.Linear(input_dim, output_dim)
        
    def forward(self, x):
        return self.lin(x)


model = LinealRegression(input_size, output_size)

# Trainnig
learning_rate = 0.01
n_iters = 80

# loss = MSE
loss = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate)

for epoch in range(n_iters):
    # predicción = forward pass
    y_pred = model(X)
    
    # pérdida
    l = loss(Y, y_pred)
    
    # grandients = backward()
    l.backward()
    
    #Actulizar los pesos
    optimizer.step()
        
    # Poner en cero los gradientes
    optimizer.zero_grad()
        
    if epoch % 2 == 0:
        #Guardamos los parametros en esta lista
        [w, b] = model.parameters()
        print(f"epoch {epoch + 1}: w = {w[0][0].item():.3f}, loos = {l:.8f}")
        
print(f"Prediction after trainig: f(5) = {model(X_test).item():.3f}")
```
