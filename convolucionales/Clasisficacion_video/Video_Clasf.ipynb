 - Luego, definiremos dos instancias de la clase para los conjuntos de datos de entrenamiento y prueba. 
 -  A continuación, definiremos dos cargadores de datos. Los cargadores de datos nos ayudarán a capturar automáticamente mini lotes del conjunto de datos durante el entrenamiento. Por ejemplo, si establecemos `batch_size=8`, los cargadores de datos devolverán minilotes (tensores) de forma `[8, 16, 3, 112, 112]`en cada iteración.

## Defining Dataset

```python
from torch.utils.data import Dataset, DataLoader, Subset
import glob
from PIL import Image
import torch
import numpy as np
import random
np.random.seed(2022)
random.seed(2022)
torch.manual_seed(2022)

class VideoDataset(Dataset):
    
    #Inicializa los valores
    def __init__(self, ids, labels, transform):      
        self.transform = transform
        self.ids = ids
        self.labels = labels
    
    #Devuelve la logitud de "ids"
    def __len__(self):
        return len(self.ids)
    
    #Devuelve los frames de un video y su respectva categoria
    def __getitem__(self, idx):
        path2imgs=glob.glob(self.ids[idx]+"/*.jpg")
        path2imgs = path2imgs[:timesteps]
        label = labels_dict[self.labels[idx]]
        frames = []
        for p2i in path2imgs:
            frame = Image.open(p2i)
            frames.append(frame)
        
        seed = np.random.randint(1e9)        
        frames_tr = []
        for frame in frames:
            random.seed(seed)
            np.random.seed(seed)
            frame = self.transform(frame)
            frames_tr.append(frame)
        if len(frames_tr)>0:
            frames_tr = torch.stack(frames_tr)
        return frames_tr, label
```

```python
# elige uno
model_type = "3dcnn"
model_type = "rnn"    

timesteps = 16

# Inicia para cada tipo de modelo
# - h, w: Alto y ancho de la imgen
# - mean: Media 
# - std: desviación estándar a lo largo del eje especificado
if model_type == "rnn":
    h, w = 224, 224
    mean = [0.485, 0.456, 0.406]
    std = [0.229, 0.224, 0.225]
else:
    h, w = 112, 112
    mean = [0.43216, 0.394666, 0.37645]
    std = [0.22803, 0.22145, 0.216989]
```

```python
import torchvision.transforms as transforms

train_transformer = transforms.Compose([
            transforms.Resize((h,w)),# Cambie el tamaño de la imagen de entrada al tamaño dado.
            transforms.RandomHorizontalFlip(p=0.5),# Voltea horizontalmente la imagen dada al azar con una probabilidad dada.  
            transforms.RandomAffine(degrees=0, translate=(0.1,0.1)),# Transformación aleatoria afín de la imagen manteniendo el centro invariante.
            transforms.ToTensor(), # Comvertimos la imgen a un tensor
            transforms.Normalize(mean, std), # Normalice una imagen de tensor con media y desviación estándar.
            ])     
```

## Defining Data Loaders

```python
# "collate_fn_r3d_18" este es una conjunto de transformaciones para nuestro modelo "3dcnn"
def collate_fn_r3d_18(batch):
    #Tomamos y almacenamos nuestros frames("imgs_batch") y nuestras estiquetas("label_batch")
    imgs_batch, label_batch = list(zip(*batch))
    
    #Los almacenamos en una lista nuestros frames(imgs_batch) y etiquetas(label_batch)
    imgs_batch = [imgs for imgs in imgs_batch if len(imgs)>0]
    label_batch = [torch.tensor(l) for l, imgs in zip(label_batch, imgs_batch) if len(imgs)>0]
    
    #Transformamos nuestras listas("imgs_batch" y "label_batch") en tensores("imgs_tensor" y "labels_tensor")
    imgs_tensor = torch.stack(imgs_batch)
    imgs_tensor = torch.transpose(imgs_tensor, 2, 1)
    labels_tensor = torch.stack(label_batch)
    
    #Devolvemos nuestros tensores
    return imgs_tensor,labels_tensor

# "collate_fn_rnn" este es una conjunto de transformaciones para nuestro modelo "rnn"
def collate_fn_rnn(batch):
    #Tomamos y almacenamos nuestros frames("imgs_batch") y nuestras estiquetas("label_batch")
    imgs_batch, label_batch = list(zip(*batch))
    
    #Los almacenamos en una lista nuestros frames(imgs_batch) y etiquetas(label_batch)
    imgs_batch = [imgs for imgs in imgs_batch if len(imgs)>0]
    label_batch = [torch.tensor(l) for l, imgs in zip(label_batch, imgs_batch) if len(imgs)>0]
    
    #Transformamos nuestras listas("imgs_batch" y "label_batch") en tensores("imgs_tensor" y "labels_tensor")
    imgs_tensor = torch.stack(imgs_batch)
    labels_tensor = torch.stack(label_batch)
    
    #Devolvemos nuestros tensores
    return imgs_tensor,labels_tensor
    
#Definimos el tamaño de nuestro mini lote
batch_size = 5

# Si nuestro tipo de modelo es un "RNN" aplicamos las transformaciones de "collate_fn_rnn"
if model_type == "rnn":
    #Crearmos nuestro conjunto de datos de Entrenamiento
    train_dl = DataLoader(train_ds, batch_size=batch_size,
                          shuffle=True, collate_fn=collate_fn_rnn)
    
    #Crearmos nuestro conjunto de datos de Testeo
    test_dl = DataLoader(test_ds, batch_size=2*batch_size,
                         shuffle=False, collate_fn=collate_fn_rnn)

# Si nuestro tipo de modelo es un "3DCNN" aplicamos las transformaciones de "collate_fn_r3d_18"
else:
    #Crearmos nuestro conjunto de datos de Entrenamiento
    train_dl = DataLoader(train_ds, batch_size=batch_size, 
                          shuffle=True, collate_fn=collate_fn_r3d_18)
    
    #Crearmos nuestro conjunto de datos de Testeo
    test_dl = DataLoader(test_ds, batch_size=2*batch_size, 
                         shuffle=False, collate_fn=collate_fn_r3d_18)          
```

## Defining Model

```python
from torch import nn

class Resnt18Rnn(nn.Module):
    def __init__(self, params_model):
        super(Resnt18Rnn, self).__init__()
        
        #Definimos los parametros de nuestros (los sacamos del diccionario "params_model")
        num_classes = params_model["num_classes"]
        dr_rate = params_model["dr_rate"]
        pretrained = params_model["pretrained"]
        rnn_hidden_size = params_model["rnn_hidden_size"]
        rnn_num_layers = params_model["rnn_num_layers"]
        
        #Cargamos el modelo "resnet18" preentrnanado
        baseModel = models.resnet18(pretrained=pretrained)
        
        #Definimos las entradas de nuestro modelo
        num_features = baseModel.fc.in_features
        
        # Inicimos nuestro modelo
        baseModel.fc = Identity()
        self.baseModel = baseModel
        
        # Agregamos una capa de "Dropout" usado el ratio(dr_rate)
        # Esta capa sirve para evitar el sobre ajuste
        self.dropout= nn.Dropout(dr_rate)
        
        # Agregamos una capa de "LSTM", Esta nos va a permitir relacionar los frames de la entrada para tener un vison global de video 
        # Y le definimos: (Numero de entradas, Numero de Salidas, Numero de capas)
        self.rnn = nn.LSTM(num_features, rnn_hidden_size, rnn_num_layers)
        
        # Agregamos una capa de "Linear" Que nos permitira clasificar nuestro video
        # Y le definimos: (Numero de entradas, Numero de Salidas)
        self.fc1 = nn.Linear(rnn_hidden_size, num_classes)
    
    def forward(self, x):
        #Definimos algunos parametros para que tengan la misma forma de "x"
        b_z, ts, c, h, w = x.shape
        
        # Definimos nuestro indice de nuestra imgen/frame actual
        ii = 0
        
        # Le pasmos al modelo base nuestra imgen/frame actual
        y = self.baseModel((x[:,ii]))
        
        # Le damos el resultado de nuestro modelo base a nuestra capa "RNN"
        # Y los guardamos en "output" y "(hn, cn)"
        output, (hn, cn) = self.rnn(y.unsqueeze(1))
        
        # Repetimos el proceso anterios para cada una de las imgen/frame de video
        for ii in range(1, ts):
            y = self.baseModel((x[:,ii]))
            out, (hn, cn) = self.rnn(y.unsqueeze(1), (hn, cn))
        
        # Le damos el resultado de nuestra capa "RNN" a nuestra capa "Dropout"
        out = self.dropout(out[:,-1])
        
        # Le damos el resultado de nuestra capa "Dropout" a nuestra capa "Linear"
        out = self.fc1(out) 
        
        # Retornamos el resultado de nuestra capa "Lineal"
        return out 
    
# Hacamos una clase que inicie nuestro modelo y ejecute nuestro "forward"
class Identity(nn.Module):
    def __init__(self):
        super(Identity, self).__init__()
    
    def forward(self, x):
        return x   
```

```python
from torchvision import models
from torch import nn

# Definimos el diccionario que contiene los parametros que nesecita nuestro modelo(RNN)
if model_type == "rnn":
    params_model={
        "num_classes": num_classes,
        "dr_rate": 0.1,
        "pretrained" : True,
        "rnn_num_layers": 1,
        "rnn_hidden_size": 100,}
    model = Resnt18Rnn(params_model)        

# Definimos el diccionario que contiene los parametros que nesecita nuestro modelo(3DCNN)
else:
    model = models.video.r3d_18(pretrained=True, progress=False)
    num_features = model.fc.in_features
    model.fc = nn.Linear(num_features, num_classes)
```

```python
# Le decimos a nuestro modelo que no ajuste sus gradientes para testearlo
with torch.no_grad():
    if model_type=="rnn":
        x = torch.zeros(1, 16, 3, h, w)
    else:
        x = torch.zeros(1, 3, 16, h, w)
    y= model(x)
    print(y.shape)
```

## Training Model

```python
from torch import optim
from torch.optim.lr_scheduler import CosineAnnealingLR, ReduceLROnPlateau
from tqdm import tqdm_notebook
import copy

loss_func = nn.CrossEntropyLoss(reduction="sum")
opt = optim.Adam(model.parameters(), lr=3e-5)
lr_scheduler = ReduceLROnPlateau(opt, mode='min',factor=0.5, patience=5,verbose=1)
os.makedirs("./models", exist_ok=True)

best_loss=float('inf')

# Obtener el ratio de aprendizaje
def get_lr(opt):
    for param_group in opt.param_groups:
        return param_group['lr']

for epoch in range(20):

    current_lr=get_lr(opt)
    #xt = caracteristicas/features de entrenamiento
    #yt = Objetivo a predecir/Taget de entrenamiento
    for xt, yt in tqdm_notebook(train_dl):

      # Ponemos el modelo en modo de entrenamiento
      model.train()

      #Movemos nuestros datos al dispositivo que usamos (CPU o GPU)
      xt, yt = xt.cuda(), yt.cuda()
      
      # Forward pass(pase hacia adelante)
      y_predicted = model(xt)
    
      # Loss(Perdida)
      loss = loss_func(y_predicted, yt)
    
      # Backward pass(pase hacia atras)
      loss.backward()
    
      # Actulizacion
      opt.step()
    
      # Vover a 0 los gradientes
      opt.zero_grad()
    
    # Resultados del Entrenamiento
    if (epoch+1) % 1 == 0:
      print(f"Resultados Entrenaiento \nepoch: {epoch+1}, loss = {loss.item():.4f}")
      
    # Ponemos el modelo en modo de entrenamiento
    model.eval()

    #Le decimos que no modifice sus gradientes
    with torch.no_grad():
      #xe = caracteristicas/features de Evalucion
      #ye = Objetivo a predecir/Taget de Evalucion
      for xe, ye in tqdm_notebook(test_dl):
        #Movemos nuestros datos al dispositivo que usamos (CPU o GPU)
        xe, ye = xe.cuda(), ye.cuda()

        # Forward pass(pase hacia adelante)
        y_predicted = model(xe)

        # Loss(Perdida)
        loss = loss_func(y_predicted, ye)

      
      #loss_history["val"].append(val_loss)
      # Resultados del Testeo
      if (epoch+1) % 1 == 0:
        print(f"Resultados Testeo \nepoch: {epoch+1}, loss = {loss.item():.4f}")

        if loss < best_loss:
            best_loss = loss
            best_model_wts = copy.deepcopy(model.state_dict())
            torch.save(model.state_dict(), path2weights)
            print("Copied best model weights!")

        lr_scheduler.step(loss)
        if current_lr != get_lr(opt):
            print("Loading best model weights!")
            model.load_state_dict(best_model_wts)
```
