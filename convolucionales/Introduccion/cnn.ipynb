```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np
```

# Hiperparámetros

```python
# Configuración del dispositivo(device)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

num_epochs = 5
batch_size = 4
learning_rate = 0.001
```

## Datos: CIFAR10

```python
# conjunto de datos tiene imágenes PILImage de rango [0, 1]. 
# Los transformamos a Tensores de rango normalizado [-1, 1]
transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

# CIFAR10: 60000 imágenes a color de 32x32 en 10 clases, con 6000 imágenes por clase
train_dataset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform)

test_dataset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size,
                                          shuffle=True)

test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=batch_size,
                                         shuffle=False)

# Guardamsos las posibles clases
classes = ('plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

```python
def imshow(img):
    img = img / 2 + 0.5  # unnormalize
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()
```

```python
# obtener algunas imágenes de entrenamiento aleatorias
dataiter = iter(train_loader)
images, labels = dataiter.next()

# mostrar imagenes
imshow(torchvision.utils.make_grid(images))
```

## Entender CNN

### Diferentes tipos de CNN

**1D CNN :** con estos, el kernel de CNN se mueve en una dirección. Las CNN 1D generalmente se usan en datos de series temporales.

**CNN 2D :** este tipo de núcleos de CNN se mueven en dos direcciones. Verá que se utilizan con el etiquetado y el procesamiento de imágenes.

**CNN 3D :** este tipo de CNN tiene un núcleo que se mueve en tres direcciones. Con este tipo de CNN, los investigadores las usan en imágenes 3D como tomografías computarizadas y resonancias magnéticas.

En la mayoría de los casos, verá CNN 2D porque se asocian comúnmente con datos de imagen. Estas son algunas de las aplicaciones para las que puede ver que se utilizan las CNN. 

La formula para calcular el tamaño de la salidad de nuestra CNN: **((Input - Filter + 2(padding) ) / Stride) + 1**

```python
# nn.Conv2d(Canales de entrada, Canales de salida, Filtro)
conv1 = nn.Conv2d(3, 6, 5)

# Esto simplifica la vision de los pixeles a una estructura de 2x2(Alto x Ancho), tomando solo los tamaños mas grandes 
# Reducuiendo asi el tamaño de la imagen a la mitad
pool = nn.MaxPool2d(2, 2)

# nn.Conv2d(Canales de entrada, Canales de salida, Filtro)
conv2 = nn.Conv2d(6, 16, 5)


print(images.shape)

# En nuestro caso seria:
# ((32 - 5 +2(0))/1) + 1 = 28(tamaño de la imagen)
X = conv1(images)
print(X.shape)

X = pool(X)
print(X.shape)

# En nuestro caso seria:
# ((14 - 5 +2(0))/1) + 1 = 10(tamaño de la imagen)
X = conv2(X)
print(X.shape)

X = pool(X)
print(X.shape)
```

## Red Neuronal(cnn)

```python
class ConvNet(nn.Module):
    def __init__(self):
        super(ConvNet, self).__init__()
        
        # nn.Conv2d(Canales de entrada, Canales de salida, Filtro)
        self.conv1 = nn.Conv2d(3, 6, 5)
        
        self.pool = nn.MaxPool2d(2, 2)
        
        # nn.Conv2d(Canales de entrada, Canales de salida, Filtro)
        self.conv2 = nn.Conv2d(6, 16, 5)
        
        #Definimos el input de esta capa de la siguiente forma 
        #nn.Linear(Canales de entrada esperados * Alto(esperado de la imagen) * Ancho(esperado de la imagen), salida)
        self.fc1 = nn.Linear(16 * 5 * 5, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10) # La salida es 10 por que hay 10 posibles clases

    def forward(self, x):
        # -> n, 3, 32, 32
        x = self.pool(F.relu(self.conv1(x)))  # -> n, 6, 14, 14 / Canales de entrada, Alto, Ancho
        x = self.pool(F.relu(self.conv2(x)))  # -> n, 16, 5, 5 / Canales de entrada, Alto, Ancho
        
        #Aqui aplanamos el tensor para que sea unidimensional
        #(Numero lotes de entrenamieto, Canales de entrada esperados * Alto de la imagen * Ancho de la imagen)
        x = x.view(-1, 16 * 5 * 5)            # -> n, 400 / Salida
        x = F.relu(self.fc1(x))               # -> n, 120 / Salida
        x = F.relu(self.fc2(x))               # -> n, 84  / Salida
        x = self.fc3(x)                       # -> n, 10  / Salida / Es 10 por que hay 10 posibles clases
        return x
    
model = ConvNet().to(device)
```

## Perdida(Loss) y optimizador


```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

## Bucle de entrenamiento

```python
n_total_steps = len(train_loader)
for epoch in range(num_epochs):
    for i, (images, labels) in enumerate(train_loader):
        # origin shape: [4, 3, 32, 32] = 4, 3, 1024
        # input_layer: 3 input channels, 6 output channels, 5 kernel size
        images = images.to(device)
        labels = labels.to(device)

        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Backward and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if (i+1) % 2000 == 0:
            print (f'Epoch [{epoch+1}/{num_epochs}], Step [{i+1}/{n_total_steps}], Loss: {loss.item():.4f}')

print('Finished Training')
PATH = './cnn.pth'
torch.save(model.state_dict(), PATH)
```

## Testeo

```python
with torch.no_grad():
    n_correct = 0
    n_samples = 0
    n_class_correct = [0 for i in range(10)]
    n_class_samples = [0 for i in range(10)]
    
    for images, labels in test_loader:
        images = images.to(device)
        labels = labels.to(device)
        outputs = model(images)
        # rendimientos máximos (valor, índice)
        _, predicted = torch.max(outputs, 1)
        n_samples += labels.size(0)
        n_correct += (predicted == labels).sum().item()
        
        for i in range(batch_size):
            label = labels[i]
            pred = predicted[i]
            if (label == pred):
                n_class_correct[label] += 1
            n_class_samples[label] += 1

    # Calculamos la precision general de nuestro modelos
    acc = 100.0 * n_correct / n_samples
    print(f'Accuracy of the network: {acc} %')

    # Mostramos la precision de nuestro modelo por clase
    for i in range(10):
        acc = 100.0 * n_class_correct[i] / n_class_samples[i]
        print(f'Accuracy of {classes[i]}: {acc} %')
```
