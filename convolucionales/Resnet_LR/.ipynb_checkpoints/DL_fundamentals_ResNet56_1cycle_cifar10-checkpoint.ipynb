```python
import numpy as np
import matplotlib.pyplot as plt

#Cosas PyTorch
import torch
torch.manual_seed(0)
from torch import optim, nn
from torch.nn import functional as F
from torchvision import transforms as T
from torchvision import models, datasets
from torch.utils.data import DataLoader, Dataset, random_split

#para usuarios de temas jupyter
from jupyterthemes import jtplot
jtplot.style()
```

```python
PATH = 'data'        # Ruta donde se van a guardar los datos
TRAIN_SIZE = 50000  # Tamaño de nuestros datos de entrenamiento
VAL_SIZE = 5000     # Tamaño de nuestros datos validación
TEST_SIZE = 5000    # Tamaño de nuestros datos de testeo
MINIBATCH_SIZE = 512# Tamaño de nuestros mini lostes

## Aplicamos trasformaciones a las imágenes de entrenamiento: 
# - volvemos las imágenes un tensor 
# - las normalizamos(con nuestra media y desviación)
# - aumentamos nuestros datos de forma random (como técnica de regularización):
#   * Ajustamos mínimamente el contraste, el brillo, la vibración el tono
#   * Hacemos una pequeña rotación
#   * Recortamos la imagen dada en el centro.
#   * Cambiamos el tamaño de salida

transform_cifar10_train = T.Compose([
                T.RandomHorizontalFlip(p=0.3),
                T.ColorJitter(brightness=0.1, contrast=0.1, hue = 0.05),
                T.RandomApply([T.RandomRotation(10), T.Resize(40), T.CenterCrop(32)], p = 0.1),
                T.ToTensor(),
                T.Normalize([0.491, 0.482, 0.447], [0.247, 0.243, 0.262])
            ])
## Aplicamos trasformaciones a las imágenes de testeo: 
# - volvemos las imágenes un tensor 
# - las normalizamos(con nuestra media y desviacion)
transform_cifar10_test = T.Compose([
                T.ToTensor(),
                            #Medias                 #Desviciones estandar
                T.Normalize([0.491, 0.482, 0.447], [0.247, 0.243, 0.262])
            ])

## Cargador de set de entrenamiento
# Cargamos y aplicamos nuestras transformaciones
cifar10_train = datasets.CIFAR10(PATH, train=True, download=True,transform=transform_cifar10_train)
# Creamos nuestros mini lotes
train_loader = DataLoader(cifar10_train, batch_size=MINIBATCH_SIZE, shuffle = True)

## Conjuntos de validación y prueba
# Cargamos y aplicamos nuestras transformaciones
test_dataset = datasets.CIFAR10(PATH, train=False, download=True, transform=transform_cifar10_test)

# Dividimos nuestros datos de validación y testeo
val_dataset, test_dataset = random_split(test_dataset, [VAL_SIZE, TEST_SIZE])

# Creamos nuestros mini lotes de validación
val_loader = DataLoader(val_dataset, batch_size=MINIBATCH_SIZE, shuffle = True)

# Creamos nuestros mini lotes de testeo
test_loader = DataLoader(test_dataset, batch_size=MINIBATCH_SIZE, shuffle = True)
```

```python
def plot_cifar10_grid():
    classes = train_loader.dataset.classes
    total_samples = 8
    plt.figure(figsize=(15,15))
    for label, sample in enumerate(classes):
        class_idxs = np.flatnonzero(label == np.array(train_loader.dataset.targets))
        sample_idxs = np.random.choice(class_idxs, total_samples, replace = False)
        for i, idx in enumerate(sample_idxs):
            plt_idx = i*len(classes) + label + 1
            plt.subplot(total_samples, len(classes), plt_idx)
            plt.imshow(train_loader.dataset.data[idx])
            plt.axis('off')
            
            if i == 0: plt.title(sample)
    plt.show()

plot_cifar10_grid() 
```

```python
def accuracy(model, loader):
    '''
    model - PyTorch model
    loader - PyTorch dataloader
    
    Returns:
    
    '''
    # Suma total de predicciones corectas
    correct = 0
    # Total de predicciones
    total = 0
    # Costo 
    cost = 0

    # Ponemos el modelo en modo de evaluación
    model.eval()
    # Trasladamos el modelo al dipositivo usado (GPU o CPU)
    model = model.to(device=device)
    
    # No actualizamos los gradientes
    with torch.no_grad():
        # Iteramos por nuestro mini lote
        for x, y in loader:
            # Movemos los datos a el dipositivo usado (GPU o CPU)
            x = x.to(device=device, dtype=torch.float32)
            y = y.to(device=device, dtype=torch.long)

            # Guardamos el prediccion del modelo
            scores = model(x)
            #Calculamos el costo
            cost += (F.cross_entropy(scores, y)).item()
            _, pred = scores.max(dim=1)
            # Si es correcta sumamos un punto en el total de correctas
            correct += (pred == y).sum()
            # Su mamos un punto en el total de predicciones echas
            total += pred.size(0)
        # Devolvemos la precisión (accuracy)
        return cost/len(loader), float(correct)/total
```

#### Buscar learning rate:

Para buscar el "ratio de aprendizaje"/"learning rate" usaremos la teoria del documento :https://arxiv.org/pdf/1803.09820.pdf

```python
def find_lr(model, optimiser, start_val = 1e-6, end_val = 10, beta = 0.99, loader = train_loader):
    n = len(loader) - 1                   # Guardamos el numero de nuestro mini lote (el menos uno por que el primero cuenta como 0)
    factor = (end_val / start_val)**(1/n) # Iniciamos nuestro factor el cual nos va a servir para averiguar nuestro ratio de aprendizaje/ learning rate
    lr = start_val                        # Iniciamos nuestro Ratio de aprendizaje
    optimiser.param_groups[0]['lr'] = lr  # Esto le permite actualizar la tasa de aprendizaje(de nuestro optimizador)
    avg_loss, loss, acc = 0., 0., 0.      # Iniciamos nuestros valores "perdida media, perdida y precision"
    lowest_loss = 0.                      # Iniciamos nuestro valor de nuestra perdida mas baja 
    batch_num = 0                         # Iniciamos nuestro valor de nuestro mini lote actual
    
    losses = []                           # Creamos una lista para guardar el progreso de nuestras "perdidas"
    log_lrs = []                          # Creamos una lista para guardar el progreso de nuestro "Ratio de aprendizaje"
    accuracies = []                       # Creamos una lista para guardar el progreso de nuestra "precisión"
    
    model = model.to(device=device)       # Configuramos nuestro dispositivo (GPU/CPU)
    
    for i, (x, y) in enumerate(loader, start=1):
        # Movemos nuestros datos a nuestro dispositivo(GPU/CPU)
        x = x.to(device = device, dtype = torch.float32)  
        y = y.to(device = device, dtype = torch.long)
        
        optimiser.zero_grad()                           # Vaciamos los gradientes para que no se acumulen
        scores = model(x)                               # Guardamos la predicción de nuestro modelo
        cost = F.cross_entropy(input=scores, target=y)  # Guardamos los resultados de nuestro modelo
        loss = beta*loss + (1-beta)*cost.item()         # Guardamos la perdida de nuestro modelo
        
        #corrección de sesgos/bias
        avg_loss = loss/(1 - beta**i)
        
        # Calculamos nuestra precision
        acc_ = ((torch.argmax(scores, dim=1) == y).sum()/scores.size(0)) 
        
        # Calculamos un promedio exponencial ponderado de nuestra precisión (con el fin de que se vea mas claramente en el grafico)
        # Tenga en cuanta que con este método puede tener un gran pico al inicio (ademas de tener una visión un poco distorsionada)
        acc = beta*acc + (1-beta)*acc_.item()
        avg_acc = acc/(1 - beta**i)
        
        #if loss is massive stop
        if i > 1 and avg_loss > 4 * lowest_loss:
            print(f'from here{i, cost.item()}')
            return log_lrs, losses, accuracies
        if avg_loss < lowest_loss or i == 1:
            lowest_loss = avg_loss
        
#        accuracies.append(acc.item())  
#        accuracies.append(acc_.item()) # Si quisiéramos guardar la precisión sin el promedio exponecial ponderado 
        accuracies.append(avg_acc)      # Si quisiéramos guardar la precisión con el promedio exponecial ponderado 
        losses.append(avg_loss)         # Guardamos nuestra perdida
        log_lrs.append(lr)              # Guardamos nuestro ratio de aprendizaje/learning rate
        
        # Propagamos el erro por la red
        cost.backward()
        optimiser.step()
        
        # Actualizamos nuestro ratio de aprendizaje/learning rate
        print(f'cost:{cost.item():.4f}, lr: {lr:.4f}, acc: {acc_.item():.4f}')
        lr = lr*factor
        optimiser.param_groups[0]['lr'] = lr
        
    return log_lrs, losses, accuracies  # Devolvemos nuestro "ratio de aprendizaje", nuestra "perdida", nuestra "precisión"
```

### Loop de Entrenamiento

```python
!mkdir models
```

```python
def train(model, optimiser, scheduler = None, epochs = 100):
    # Configuramos nuestro dispositivo (GPU/CPU)
    model = model.to(device = device)
    
    val_loss_history = []   # Creamos una lista para guardar el progreso de nuestra "perdida" en el testeo
    train_loss_history = [] # Creamos una lista para guardar el progreso de nuestra "perdida" en el entrenamiento
    val_acc_history = []    # Creamos una lista para guardar el progreso de nuestra "precisión" en el testeo
    train_acc_history = []  # Creamos una lista para guardar el progreso de nuestra "precisión" en el entrenamiento
    lrs = []                # Creamos una lista para guardar el progreso de nuestro "Ratio de aprendizaje"
    train_cost = 0.         # Iniciamos nuestro costo de entrenamiento
    val_cost = 0.           # Iniciamos nuestro costo de evaluación
    train_cost_acum = 0     # Iniciamos el costo de entrenamiento acumulado

    for epoch in range(epochs):
        train_correct_num  = 0  # Iniciamos nuestro numero de respuestas correctas
        train_total = 0         # Iniciamos nuestro numero de intentos
        train_cost_acum = 0     # Iniciamos nuestro costo total de entrenamiento

        # Entrenamos con nuestros mini lotes
        for mb, (x, y) in enumerate(train_loader, start=1):
            # Ponemos nuestro modelo en modo de entrenamiento
            model.train()

            # Iniciamos nuestros datos en nuestro dispositivo (CPU/GPU)
            x = x.to(device=device, dtype=torch.float32)
            y = y.to(device=device, dtype=torch.long)
            
            # Guardamos el resultado de nuestro modelo
            scores = model(x)
            
            # Calculamos nuestro costos
            cost = F.cross_entropy(input=scores, target=y)
            
            optimiser.zero_grad() # Vaciamos nuestro gradientes
            cost.backward()       # Propagamos el error por nuestra red
            optimiser.step()      # Optimizamos nuestro modelo
            
            #Si usamos "scheduler" ajustamos nuestro "scheduler"
            if scheduler: 
              scheduler.step()
                
            train_correct_num += (torch.argmax(scores, dim=1) == y).sum() #Sumamos nuestro numero de respuestas correctas
            train_total += scores.size(0)                                 #Sumamos nuestro numero de intentos
            train_cost_acum += cost.item()                                #Sumamos nuestro costo de entrenamiento
            train_acc = float(train_correct_num)/train_total              #Obtenemos nuestro precisión de entrenamiento
#             train_cost = train_cost_acum/mb
            val_cost, val_acc = accuracy(model, val_loader)               #Obtenemos nuestro costo y presicion de evalucion

            val_loss_history.append(val_cost)           # Guardamos nuestro costo de evaluación
            train_loss_history.append(cost.item())      # Guardamos nuestro costo de entrenamiento
            val_acc_history.append(val_acc)             # Guardamos nuestra precisión de evaluación
            train_acc_history.append(train_acc)         # Guardamos nuestro precisión de entrenamiento
#             lrs.append(scheduler.get_last_lr()[0])
            lrs.append(optimiser.param_groups[0]["lr"]) # Guardamos nuestros ratio de aprendizaje/learning rate
        
        #f'last lr: {scheduler.get_last_lr()[0]:6f},
        
        # Cada 10 épocas guardamos nuestro modelo
        if epoch%10 ==0:
            torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizaer_state_dict': optimiser.state_dict(),
                'loss': cost,
            },f'models/check{epoch}.pt')
        
        # Calculamos nuestra precisión de entrenamiento final 
        train_acc = float(train_correct_num)/train_total
        # Calculamos nuestro costo de entrenamiento final
        train_cost = train_cost_acum/len(train_loader)
        
        print(f'Epoch:{epoch}, train cost: {train_cost:.6f}, val cost: {val_cost:.6f},'
              f' train acc: {train_acc:.4f}, val acc: {val_acc:4f}, total: {train_total},'
              f' lr: {optimiser.param_groups[0]["lr"]:.6f}')
        
    
    # Regresamos nuestro historial de perdida del entrenamiento y evaluación
    # Regresamos nuestro historial de precisión del entrenamiento y evaluación
    # Regresamos nuestro ratio de aprendizaje/learning rate
    return train_loss_history, val_loss_history, train_acc_history, val_acc_history, lrs
```

### ResNet56

```python
# Creamos una variable que almacene una capa convulcional, para reutilizarla y no tener que repetir código debido a que vamos a usar siempre los mismos valores de "kernel_size"y"padding"
conv_k_3 = lambda channel1, channel2, stride: nn.Conv2d(channel1, channel2, stride = stride, kernel_size=3, padding=1)

class residual_block(nn.Module):
    '''
    Esta clase es para crear el "residual block" basico (de Resnet),
    in_channel: Permite definir el numero de canales de entrada , "!=out_channel" solo "w"
    out_channel: Permite definir los canales de salida, usualmente = el input
    stride: cambios de zancada al cambiar los tamaños de activación
    change_size: Bandera para indicar si las activaciones estan dentro de los bloques residuales normales
    '''
    def __init__(self, in_channel, out_channel, stride=1, change_size = True):
        super().__init__()
        self.conv1 = conv_k_3(in_channel, out_channel, stride)
        self.bn1 = nn.BatchNorm2d(out_channel)
        self.conv2 = conv_k_3(out_channel, out_channel, 1)
        self.bn2 = nn.BatchNorm2d(out_channel)
        # Para cambiar el tamaño del mapa de activación
        self.change_size = change_size
        if change_size:
            self.residual = nn.Sequential(nn.Conv2d(in_channel, 
                                                    out_channel, 
                                                    kernel_size=1, # Este es el bloque de uno por uno que nos permite cambiar el tamaño
                                                    stride=stride),
                                         nn.BatchNorm2d(out_channel)
                                         )      
    def forward(self, x):
        # Aqui creamos la informacion residual/skip connections
        # Si no hacemos cambio de tamaño pasamos la informacion tal cual, si hecmos cambio de tamaño cambiamos nuestra informacion con la funcion "residual"
        identity = x if not self.change_size else self.residual(x)
        
        # Aqui creamos nuestro bloque
        y = F.relu(self.bn1(self.conv1(x))) # Primero pasamos nuestra información por una capa "capa "conv_k_3"(conv1), normalizamos "BatchNorm2d"(bn1), Activación ReLu 
        y = self.bn2(self.conv2(y))         # Después pasamos por otra capa capa "conv_k_3"(conv2) y por 
        
        # Al final aplicamos un capa "Relu"
        return F.relu(y)
```

```python
class ResNet56(nn.Module):
    def __init__(self, n=9, num_classes=10):
        super().__init__()
        # Creamos una capa "conv_k_3"
        self.conv1 = conv_k_3(3, 16, stride = 1)
        # Una capa de normalización, con input de "16"
        self.bn1 = nn.BatchNorm2d(16)
        
        #Creamos un bloque que reciba 16 canales de entrada, con canales 16 de salida, stride de 1 y no cambiamos su tamaño
        #Esto crearia un total de 18 capas
        self.block1 = self.create_block(n=9, in_channel=16, 
                                        out_channel=16, stride=1, 
                                        change_size=False)
        
        # Creamos un bloque que reciba 16 canales de entrada, con canales 36 de salida, stride de 2(para reducir la imagen, pero aumentar sus canales de salida) y si cambiamos su tamaño
        # Esto crearia un total de 18 capas
        self.block2 = self.create_block(n=9, in_channel=16, 
                                        out_channel=32, stride=2)
        
        # Creamos un bloque que resiva 32 canales de entrada, con canales 64 de salida, stride de 2(para reducir la imagen, pero aumentar sus canales de salida) y si cambiamos su tamaño
        # Esto crearía un total de 18 capas
        self.block3 = self.create_block(n=9, in_channel=32, 
                                        out_channel=64, stride=2)
        
        # Creamos una capa "full connection" que reciba 64 canales de entrada, y de salida tenga la cantidad de posibles clases(10)
        self.fc = nn.Linear(64, num_classes)

    def create_block(self, n, in_channel, out_channel, stride, change_size=True):
        # Iniciamlizamos una lista con un objeto de tipo residual block
        block = [residual_block(in_channel, out_channel, stride, change_size=change_size)]

        # Hamcemos n-1 por que queremos crear 18 capas y el primer bloque de arriba ya son dos bloques
        # Si n = 9, n*2+2=20. Pero si n = 9-1, n*2+2=18
        for i in range(n-1):
            # Aquí establecemos que nuestro numero de canales de entra y salida sean iguales(esto debe seguir así si "change_size=False")
            block.append(residual_block(out_channel, out_channel, stride=1, change_size=False))
        # Regresamos una red neuronal secuencial de la lita block (el * indica que el nuemro de argumentos puede ser ser variable)
        return nn.Sequential(*block)   
        
    def forward(self, x):
        # Primero pasmos nuestra información por una capa "conv_k_3"(conv1), la normalizamos "BatchNorm2d"(bn1) y una capa de activación "Relu"
        y = F.relu(self.bn1(self.conv1(x)))

        # Despues creamos el bloque uno(block1), sobre el, el bloque dos (block2), y sobre ese el bloque tres (block3)
        y = self.block3(self.block2(self.block1(y)))
        
        # Creamos un "adaptive_avg_pool2d" en vez de una típica capa flaten para tener 64 elementos
        y = F.adaptive_avg_pool2d(y, 1)
        
        # Cambiamos la forma de nuestro tensor que tiene los datos para que sea la primera dimensión(64) por el numero de columnas(1) 64x1
        return self.fc(y.view(y.size(0), -1))
```

```python
# Instanciamos un objeto de esta clase
model_resnet56 = ResNet56()
```

```python
# model_resnet56 = ResNet56()
optimiser_resnet56 = torch.optim.SGD(model_resnet56.parameters(),
                                     # usamos este "lr" y "momentum" solo para poner algun valor inicial
                                     #sin embargo no importa por que nuestro "scheduler" lo va a ajustar 
                                     lr=0.1, momentum=0.95,
                                     # Usamos este metodo de regularizacion(weight_decay) por que senos menciona en el documento de 
                                     #"one cycle policy", ademas se nos menciona que 1e-4 suele funcionar bien 
                                     weight_decay=1e-4)
```

```python
# Usamos nuestra función "find_lr", ponemos nuestros valores "start_val" y "end_val" de forma aleatoria, intentando claro que allá un amplio rango entre estos dos
lg_lr, losses, accuracies = find_lr(model_resnet56, optimiser_resnet56,
                                    start_val=1e-6, end_val=10)
```

```python
model_resnet56 = ResNet56()
optimiser_resnet56 = torch.optim.SGD(model_resnet56.parameters(),
                                     lr=0.1, momentum=0.95,
                                     weight_decay=1e-4)
epochs = 50
scheduler = torch.optim.lr_scheduler.OneCycleLR(optimiser_resnet56, 
                                                max_lr=2e-1, #Aqui elegimos un valor ligeramente menor al "lr" que nos dio mejor resultado 
                                                steps_per_epoch=len(train_loader), 
                                                epochs = epochs, pct_start=0.43, 
                                                div_factor=10, 
                                                final_div_factor=1000, 
                                                three_phase=True, verbose=False
                                            )
train_loss_history, val_loss_history, train_acc_history, val_acc_history, lrs = train(
                                model_resnet56, 
                                optimiser=optimiser_resnet56,
                                scheduler=scheduler,
                                epochs = epochs
                                )
```
